```python
import matplotlib.pyplot as plt
import numpy as np

# Create figure and subplots
fig, ax = plt.subplots(1, 3, figsize=(15, 5))

# Define vectors
vector_A = np.array([1, 1])  # Reference vector
vector_B1 = np.array([0.8, 0.9])  # Similar vector (cos  1)
vector_B2 = np.array([-1, 1])  # Orthogonal vector (cos  0)
vector_B3 = np.array([-1, -1])  # Opposite vector (cos  -1)

# Define plotting function
def plot_vectors(ax, v1, v2, title):
    ax.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Vector A')
    ax.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='red', label='Vector B')
    ax.set_xlim(-1.5, 1.5)
    ax.set_ylim(-1.5, 1.5)
    ax.axhline(0, color='black', linewidth=0.5, linestyle='--')
    ax.axvline(0, color='black', linewidth=0.5, linestyle='--')
    ax.set_title(title)
    ax.legend()
    ax.grid()

# Plot vectors
plot_vectors(ax[0], vector_A, vector_B1, "Cosine  1 (Similar)")
plot_vectors(ax[1], vector_A, vector_B2, "Cosine  0 (Orthogonal)")
plot_vectors(ax[2], vector_A, vector_B3, "Cosine  -1 (Opposite)")

# Adjust layout and show plot
plt.tight_layout()
plt.show()
```
